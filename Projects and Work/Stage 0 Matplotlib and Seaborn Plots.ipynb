## Random Data used for practice problems (Needs to be ran)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Dataset 1: simple weekly study data
study = pd.DataFrame({
    "week": [1, 2, 3, 4, 5, 6, 7, 8],
    "python_hours": [3, 4, 5, 5, 7, 8, 8, 10],
    "math_hours": [2, 2, 3, 4, 4, 5, 6, 6],
    "practice_score": [55, 60, 63, 67, 72, 76, 80, 85]
})

# Dataset 2: benchmark timing data
benchmarks = pd.DataFrame({
    "problem_size": [100, 200, 400, 800, 1600, 3200],
    "cpu_time_ms": [1.2, 2.8, 6.5, 14.2, 32.8, 74.0],
    "gpu_time_ms": [4.0, 4.3, 5.1, 6.8, 10.9, 18.5]
})

# Dataset 3: quantum circuit experiment results
quantum_results = pd.DataFrame({
    "circuit": ["Bell", "Bell", "GHZ", "GHZ", "QFT", "QFT", "Grover", "Grover"],
    "backend": ["simulator", "hardware", "simulator", "hardware", "simulator", "hardware", "simulator", "hardware"],
    "success_rate": [0.99, 0.86, 0.97, 0.78, 0.95, 0.72, 0.93, 0.69],
    "depth": [3, 3, 7, 7, 12, 12, 18, 18],
    "shots": [1024, 1024, 1024, 1024, 2048, 2048, 2048, 2048]
})

# Dataset 4: larger plotting dataset
training = pd.DataFrame({
    "day": list(range(1, 16)),
    "topic": [
        "Python", "Python", "Python", "Matplotlib", "Matplotlib",
        "Seaborn", "Seaborn", "NumPy", "NumPy", "Pandas",
        "Pandas", "Algorithms", "Algorithms", "HPC", "HPC"
    ],
    "hours": [1.0, 1.5, 2.0, 1.0, 2.5, 1.5, 2.0, 2.5, 3.0, 2.0, 2.5, 1.5, 2.0, 2.5, 3.5],
    "confidence": [2, 3, 4, 3, 5, 4, 5, 5, 6, 5, 6, 4, 5, 5, 7],
    "quiz_score": [50, 58, 65, 62, 72, 70, 75, 78, 82, 76, 84, 73, 80, 81, 88]
})
```

# Practice Problem 1: Basic Line Plot

Use the study dataframe to make a line plot showing the week on the x-axis and python_hours on the y-axis. Make a title, x-axis label, y-axis label, and have markers on each point.

```python
fig, ax = plt.subplots()
ax.plot(study['week'], study['python_hours'], marker='o')

ax.set_title("Python Hours per Week")
ax.set_xlabel("Week")
ax.set_ylabel("Hours")
ax.grid(True, alpha = 0.25)

plt.show()
```

# Practice Problem 2: Basic Bar Graph

Use the study dataframe to make bar graph showing python_hours for each week to show which week had the most python studying time.

```python
fig, ax = plt.subplots()
ax.bar(study['week'], study['python_hours'])

ax.set_title("Python Hours per Week")
ax.set_xlabel('Week')
ax.set_ylabel('Hours')

plt.show()
```

The week with the most hours studied is week 8.

# Practice Problem 3: Basic Scatter Plot

Use the study dataframe to make a scatter plot showing practice test scores on the x-axis and practice test scores on the y axis to see if there is a correlation between the two.

```python
fig, ax = plt.subplots()
ax.scatter(study['python_hours'], study['practice_score'], marker = 'o')

ax.set_title("Python Practice Test Scores vs Hours Studied")
ax.set_xlabel('Hours Studied')
ax.set_ylabel('Score')
ax.grid(True, alpha = 0.25)

plt.show()
```

There is a positive correlation between the nuer of hours studied and practice test scores

# Practice Problem 4: Combining Two Line Plots into One Graph

Using 'study', plot python_hours and math_hours over 'week'. Make sure that the graph has two separete lines, a legend, a title, and axis labels.

```python
fig, ax = plt.subplots()
ax.plot(study['week'], study['python_hours'], marker = 'o', label = 'Python')
ax.plot(study['week'], study['math_hours'], marker = 'x', label = 'Math')

ax.set_title('Math and Python Hours Studied')
ax.set_xlabel('Week')
ax.set_ylabel('Hours')
ax.legend()
ax.grid(True, alpha = 0.25)

plt.show()
```

# Practice Problem 5: CPU vs GPU Timing Line Plot

Using 'benchmarks', plot cpu_time_ms and gpu_time_ms vs problem size. Both lines should be on the same graph. Make sure to add a legend and a grid as well as titles for everything. At small problem sizes, which is faster? At large problem sizes, which is faster?

```python
fig, ax = plt.subplots(figsize = (8, 5))
ax.plot(benchmarks['problem_size'], benchmarks['cpu_time_ms'], marker = 'o', label = 'CPU')
ax.plot(benchmarks['problem_size'], benchmarks['gpu_time_ms'], marker = 'x', label = 'GPU')

ax.set_title('GPU and CPU Timing vs Problem Size (Lower is Better)')
ax.set_xlabel('Problem Size')
ax.set_ylabel('Milliseconds')
ax.legend()
ax.grid(True, alpha = 0.25)

plt.show()
```

At small problem sizes, the CPU has faster timing than the GPU. At large problem sizes, the GPU has faster timing than the CPU. Another thing to note is that the GPU has startup overhead and so the graph does not fully paint the picture of pure speed between the CPU and GPU.

# Practice Problem 6: Seaborn Line Plot

Using 'study', plot a line plot of python_hours over 'week'. Add a title, markers, and label axes.

```python
fig, ax = plt.subplots(figsize = (7,4))
sns.lineplot(
    data=study,
    x='week',
    y='python_hours',
    marker='o',
    errorbar=None,
    ax=ax
)

ax.set(title='Python Hours Studied Per Week', xlabel='Week', ylabel='Hours')
plt.show()
```

# Practice Problem 7: Seaborn Bar Plot

# Practice Problem 9: Combining Two Plots into One Graph using Seaborn

Using 'study', plot python_hours and math_hours over 'week'. Make sure that the graph has two separete lines, a title, and axis labels.

```python
fig , ax = plt.subplots(figsize=(7,4))
sns.lineplot(
    data=study,
    x='week',
    y='python_hours',
    errorbar=None,
    ax=ax
)

sns.lineplot(
    data=study,
    x='week',
    y='math_hours',
    errorbar=None,
    ax=ax
)

sns.set_style('darkgrid')
ax.set(title='Python and Math Hours Studied over Weeks', xlabel='Week', ylabel='Hours')
plt.show()
```

# Practice Problem 10: Grouped Average Bar Chart

Using 'training', calculate the average 'quiz_score' for each 'topic'. Then make a bar chart of the averages.

```python
topic_avg = training.groupby('topic')['quiz_score'].mean().reset_index()

fig, ax = plt.subplots(figsize=(7,5))
plt.bar(topic_avg["topic"], topic_avg["quiz_score"])

ax.set_title('Average Quiz Scores')
ax.set_xlabel('Topic')
ax.set_ylabel('Average Score')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Practice Problem 14: Making a Professional Benchmark Plot

Using 'benchmarks', make a polished graph comparing CPU and GPU time.

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(benchmarks['problem_size'], benchmarks['cpu_time_ms'], marker='o', label='CPU')
ax.plot(benchmarks['problem_size'], benchmarks['gpu_time_ms'], marker='x', label='GPU')

ax.set_title('GPU vs CPU Timing',fontsize=16)
ax.set_xlabel('Problem Size',fontsize=12)
ax.set_ylabel('Time (ms)',fontsize=12)
ax.legend(fontsize=12)

plt.tight_layout()
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
